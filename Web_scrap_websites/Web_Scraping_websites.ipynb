# Web Scraping

```python
# Import all required packages
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

### URL: https://tsmliberia.com/category/fact-check/
### Get all links to all articles on the website

```python
alls = []

for i in range(1,16):
    new_url = f"https://tsmliberia.com/category/fact-check/page/{i}/"
    print(new_url)
    
    r = requests.get(url = new_url)#, proxies=proxies)
    content = r.text
    soup = BeautifulSoup(markup=content, features='html.parser')
    for d in soup.findAll('div', attrs={'class':'td-block-span4'}):
        # print(d)
        block = d.find('h3', attrs={'class':'entry-title td-module-title'})
        block_title = block.text
        print(block_title)
        href = d.find('a', href= True).get('href')
        print(href)
    
        all1=[]

        all1.append(block_title)
        all1.append(href)

        alls.append(all1)
        
    # Navigate to the next page
    print("Navigating to Next Page")
```

```python
len(alls)
```

```python
# Save the article title and link to the article in a dataframe
df = pd.DataFrame(alls,columns=['title','link'])
df.to_csv('fact_check.csv', index=False, encoding='utf-8')
```

### Scrape `title`, `author name`, `content`, `publication date`, `link` of each article

```python
# Scrape all the text from each article
# Save it in a text file with the title as its name
details = []
df = pd.read_csv('fact_check.csv')
for i in df.iterrows():
    
    # Create a list for each article detail
    detail = []
    # print(i[1].link)

    # get the title from the CSV
    title = i[1].title

    # Get the link from the CSV
    link = i[1].link

    # Get the content of each link
    response = requests.get(link)
    r = response.text
    soup = BeautifulSoup(markup=r, features='html.parser')
    
    # Append title into the detail list
    detail.append(title)
    print(title)

    # Get the author name 
    author_name = soup.find('div', attrs={'class':'td-post-author-name'}).find(name = 'a').getText()
    detail.append(author_name)
    print(author_name)

    # Get the content of the article
    content = []
    for d in soup.findAll('div', attrs={'class':'td-post-content td-pb-padding-side'}):
        content.append(d.text)
    
    # Combine all sentences inside the content list
    text_of_article = ' '.join(content)
    print(text_of_article)

    # Append text into detail list
    detail.append(text_of_article)

    # get the publication date and append into detail list
    publication_date = soup.find('span', attrs={'class':'td-post-date'}).text
    detail.append(publication_date)
    print(publication_date)

    # Append the link into the detail list
    detail.append(link)
    print(link)

    # Append each article details into a bigger list
    details.append(detail)

```

```python
# Save it into a dataframe and csv file
df_content = pd.DataFrame(details,columns=['title','publisher_name', 'content', 'publication_date', 'link'])
df_content.to_csv('data_collection.csv', index=False, encoding='utf-8')
```
